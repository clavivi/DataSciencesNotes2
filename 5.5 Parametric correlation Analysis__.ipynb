### Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec=cars['qsec']
wt = cars['wt']

pearsonr_coefficient, pvalue = pearsonr(mpg, hp)
print('PearsonR correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, pvalue = pearsonr(mpg, qsec)
print('PearsonR correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

#### weak correlation between mpg and qsec

```python
pearsonr_coefficient, pvalue = pearsonr(mpg, wt)
print('PearsonR correlation Coefficient %0.3f'% (pearsonr_coefficient))
```

#### strong negative correlation between weight and mpg
